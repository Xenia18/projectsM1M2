# Création de la variable cible

Pour une répartiton égalitaire des classes, la division de la variables cible returns se fair par percentile 33% et 66%, respectivement 1/3 et 2/3 des données.

```python
#création de la variable target
#suppression des observations avec des valeurs manquantes dans la variable cible returns
total=total.dropna(axis=0,subset=["returns"])
import numpy as np
#calcul des percentiles
p33=np.percentile(total['returns'],33)  
p66=np.percentile(total['returns'],66)              
total['returns_classes']=pd.cut(total['returns'],bins=[min(total['returns'])-1,p33,p66,max(total['returns'])+1], labels=[0, 1, 2])
```

```python
print("percentile 33% : ", p33)
print("percentile 66% : ", p66)
```

```python
#fréquence of the classes, verification
total['returns_classes'].value_counts().plot(kind='barh')
```
